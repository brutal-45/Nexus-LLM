# Getting Started with Nexus-LLM

This notebook walks you through the basics of Nexus-LLM, from installation to your first chat interaction.

## Prerequisites

- Python 3.9+
- CUDA-capable GPU (recommended) or CPU
- At least 16GB RAM

## 1. Installation

```python
# Install Nexus-LLM
!pip install nexus-llm

# Verify installation
import nexus_llm
print(f"Nexus-LLM version: {nexus_llm.__version__}")
```

## 2. Your First Chat

```python
from nexus_llm import InferenceEngine, Conversation

# Initialize the engine
engine = InferenceEngine(
    model_name="nexus-7b-chat",
    device="auto",
)

# Create a conversation
conversation = Conversation(
    system_prompt="You are a helpful AI assistant. Be concise and accurate."
)

# Send a message
conversation.add_user_message("What is the meaning of life?")
response = engine.chat(conversation)

print(f"Assistant: {response.text}")
print(f"\nTokens: {response.token_count} | Time: {response.elapsed_time:.2f}s")
```

## 3. Multi-Turn Conversation

```python
# Continue the conversation with context
conversation.add_user_message("Can you elaborate on that?")
response = engine.chat(conversation)

print(f"Assistant: {response.text}")

# Check conversation stats
print(f"\nTotal turns: {conversation.num_turns}")
print(f"Total tokens: {conversation.total_tokens}")
```

## 4. Adjusting Generation Parameters

```python
# Low temperature = more deterministic, focused
response_focused = engine.chat(
    conversation,
    temperature=0.1,
    top_p=0.9,
    max_tokens=200,
)
print(f"Focused (temp=0.1): {response_focused.text[:200]}...")

# High temperature = more creative, diverse
conversation_creative = Conversation(system_prompt="You are a creative writer.")
conversation_creative.add_user_message("Write a haiku about coding.")
response_creative = engine.chat(
    conversation_creative,
    temperature=0.9,
    top_p=0.95,
)
print(f"\nCreative (temp=0.9): {response_creative.text}")
```

## 5. Streaming Responses

```python
# Stream tokens in real-time
conversation_stream = Conversation(system_prompt="You are a storyteller.")
conversation_stream.add_user_message("Tell me a short story about a time traveler.")

print("Assistant: ", end="")
for chunk in engine.chat_stream(conversation_stream, iterate=True):
    print(chunk.token, end="", flush=True)
    if chunk.is_final:
        print()  # newline
```

## 6. Saving and Loading Conversations

```python
# Save a conversation
conversation.save("my_first_chat.json")
print("Conversation saved!")

# Load it back
loaded_conversation = Conversation.load("my_first_chat.json")
print(f"Loaded conversation with {loaded_conversation.num_turns} turns")

# Inspect the message history
for msg in loaded_conversation.messages:
    print(f"[{msg.role}]: {msg.content[:80]}...")
```
